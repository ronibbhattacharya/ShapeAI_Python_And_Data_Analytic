```python
import pandas as pd 
import numpy as np
```

```python
df= pd. Dataframes(pd.read_csv('/content/train.csv'))
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
# df.isnull().sum() returns a pandas series with columns name as the label index
# And total count of null values in the columns as its value 
# And we are storing only those columns which have more than 35% of data
drop_col= df.isnull().sum()[df.isnull().sum()>(35/100*df.shape[0])]
drop_col
```

```python
drop_col.index
```

```python
df.drop(drop_col.index,axis=1,inplace= True) 
df.isnull().sum()
```

```python
df.fillna(df.mean(),inplace= True)
df.isnull().sum()
```

```python
df['Embarked'].describe()
```

```python
df['Embarked'].fillna('S',inplace= True)
df.isnull().sum()
```

```python
df.corr()
```

```python
df['FamilySize']=df['Sibsp']+['Parch']
df. drop(['Sibsp', 'Parch'],axis=1,inplace= True) 
df.corr()
```

```python
df['Alone'] = [0 if df['FamilySize'][i]>0 else 1 for i in df.index
df.head()
```

```python
df.groupby(['Alone'])['Survived '].mean()
```

```python
df[['Alone ','Fare']].corr()
```
